```python
def emp(name,age):
    #doc string
    ''' 
    this is function is take 2 aru 
    one is name with the string data type
    and another one id age with the int data type
    '''
    print(f"The name of employee is {name}")
    print(f"The age of employee is {age}")
    
```

```python
emp("madhav",23)
emp(23,"madhav")
```

```python
emp(name = "madhav",age = 90)
emp(age = 90,name = "madhav")
```
