```python
# OOP — Classes — every LangChain component is a class
# self => this => represent the current instance
class LLMClient:
    def __init__(self, model: str, temperature: float = 0.7):
        self.model       = model
        self.temperature = temperature
        self.call_count  = 0

    def complete(self, prompt: str) -> str:
        self.call_count += 1
        # Real version comes in Module 6
        return f"[{self.model}] Response #{self.call_count}"

    def __repr__(self) -> str:
        return f"LLMClient(model={self.model}, calls={self.call_count})"

client = LLMClient("gpt-4o")
print(client.complete("What is RAG?"))
print(client.complete("Explain LangChain"))
print(client)

# Has-A : declare an object of a class into another class
# IS - A : inheritance
```

### Task 3 — Extend LLMClient
  
> *'Add a method to `LLMClient` called `get_summary()` that returns a string like: `"gpt-4o made 3 calls at temperature 0.7"`. Then create two different clients and call each 1–2 times, and print their summaries.'*

```python

# Add get_summary() to LLMClient that returns:
# "gpt-4o made 3 calls at temperature 0.7"

# Then:
# 1. Create two clients with different models
# 2. Make 1-2 calls each
# 3. Print both summaries

```

```python

class LLMClient:
    def __init__(self, model: str, temperature: float = 0.7):
        self.model       = model
        self.temperature = temperature
        self.call_count  = 0

    def complete(self, prompt: str) -> str:
        self.call_count += 1
        return f"[{self.model}] Response #{self.call_count}"

    def get_summary(self) -> str:
        return f"{self.model} made {self.call_count} calls at temperature {self.temperature}"

# Two different clients
gpt    = LLMClient("gpt-4o",  temperature=0.7)
claude = LLMClient("claude",  temperature=0.3)

gpt.complete("What is RAG?")
gpt.complete("Explain LangChain")
claude.complete("What is a vector DB?")

print(gpt.get_summary())
print(claude.get_summary())
```

## 5 — Async / Await

> *'This is the single most important Python concept for this course. Every LLM call, every agent, every FastAPI endpoint is async. Pay attention.'*

```python
#  PROBLEM  — sync is too slow for AI apps
import time

def sync_llm_call(model: str) -> str:
    time.sleep(1)             # simulates 1s API response
    return f"{model}: done"

start = time.time()
r1 = sync_llm_call("gpt-4o")
r2 = sync_llm_call("claude")
r3 = sync_llm_call("gemini")
print(f"Sequential: {time.time()-start:.1f}s   too slow")
print("In a real app with 5 agent tools each taking 2s = 10s wait. Unacceptable.")
```

```python
# THE SOLUTION — concurrent with asyncio
import asyncio, time

async def async_llm_call(model: str, prompt: str) -> str:
    await asyncio.sleep(1)         # same 1s — but doesn't block others
    return f"{model}: '{prompt[:25]}'"

async def run_all():
    start = time.time()

    results = await asyncio.gather(       # ALL THREE run at the same time
        async_llm_call("gpt-4o",  "What is RAG?"),
        async_llm_call("claude",  "Explain LangChain"),
        async_llm_call("gemini",  "What is a vector DB?"),
    )

    for r in results:
        print(f"   {r}")
    print(f"\nTime: {time.time()-start:.1f}s instead of 3.0s")

await run_all()   # Jupyter: await directly. In .py files: asyncio.run(run_all())
```

###  Task 4 — Parallel Tool Runner

> *'Write an async function `search(query)` that simulates a 1.5s web search and returns `"Results for: {query}"`. Then run THREE different searches concurrently using `asyncio.gather` and print the total time. It should finish in ~1.5s, not ~4.5s.'*

```python


# Write: async def search(query: str) -> str
#   - 1.5s simulated delay
#   - returns "Results for: {query}"

# Then run these 3 concurrently and print total time:
#   "What is LangChain?"
#   "How does RAG work?"
#   "What is a vector database?"

# Target: ~1.5s total, not ~4.5s
```

```python
# SOLUTION

import asyncio, time

async def search(query: str) -> str:
    await asyncio.sleep(1.5)
    return f"Results for: {query}"

async def run_searches():
    start = time.time()

    results = await asyncio.gather(
        search("What is LangChain?"),
        search("How does RAG work?"),
        search("What is a vector database?"),
    )

    for r in results:
        print(f"   {r}")
    print(f"\nTotal: {time.time()-start:.1f}s  ← concurrent!")

await run_searches()
```

## 6 - Decorators

```python
import time, functools

# A decorator adds behaviour to a function without changing it
def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start  = time.time()
        result = func(*args, **kwargs)
        print(f"  [{func.__name__}] {time.time()-start:.2f}s")
        return result
    return wrapper

@timer                          #  this is the decorator
def call_llm(prompt: str) -> str:
    time.sleep(0.3)
    return f"Response to: {prompt}"

result = call_llm("What is LangChain?")
print(result)
```

### Task 5 - Retry Decorator

> *'Write a `@retry(max_attempts=3)` decorator. If the function raises an exception, wait 0.5s and try again. If all attempts fail, raise the last error. Test it on a function that randomly fails 50% of the time.'*

```python
# Write a retry(max_attempts) decorator that:
#   - catches any Exception
#   - waits 0.5s between attempts
#   - raises the error if all attempts fail
```

```python
# Test with this function (randomly fails):
import random
#@retry(max_attempts=3)
def flaky_llm_call(prompt):
    print(random.random())
    if random.random() < 0.5:
        raise ConnectionError("API timeout")
    return f"Response to: {prompt}"

try:
    result = flaky_llm_call("What is RAG?")
    print(f"{result}")
except Exception as e:
    print(f"All attempts failed: {e}")

```

```python
# SOLUTION

import time, functools, random

def retry(max_attempts=3, delay=0.5):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            last_error = None
            print("args :",args)
            print("kwargs :",kwargs)
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    last_error = e
                    print(f"  Attempt {attempt} failed: {e}")
                    if attempt < max_attempts:
                        time.sleep(delay)
            raise last_error
        return wrapper
    return decorator

@retry(max_attempts=3)
def flaky_llm_call(prompt: str) -> str:
    if random.random() < 0.5:
        raise ConnectionError("API timeout")
    return f"Response: {prompt}"

try:
    #result = flaky_llm_call("What is RAG?")
    result = flaky_llm_call(prompt="What is RAG?")
    print(f"{result}")
except Exception as e:
    print(f" All attempts failed: {e}")

# Why this matters: LLM APIs have rate limits and timeouts.
# You'll wrap every real API call with retry in production.
```

## 7 — Generators & Streaming

>  *'When ChatGPT shows words appearing one by one — that's a generator. Every streaming LLM response is a generator. This is how your Module 11 chat UI will work.'*

```python
# Generator — yields one value at a time instead of returning everything at once
def stream_response(text: str):
    for word in text.split():
        yield word + " "

# Consume it — this is what Streamlit will do in Module 11
response = "RAG helps LLMs answer questions using your own documents and data"

print("Streaming: ", end="")
for token in stream_response(response):
    print(token, end="", flush=True)
    time.sleep(1)

print("\n Stream complete")
```

### Task 6 — Streaming Word Counter

> *'Write a generator `stream_with_count(text)` that yields each word one at a time AND prints a running count of how many tokens have been streamed so far. Format: `[token 1] Hello`, `[token 2] world` etc.'*

```python

# Write: stream_with_count(text: str)
# A generator that yields each word AND prints:
#   [token 1] Hello
#   [token 2] world
#   [token 3] !
#   --- 3 tokens streamed ---

text = "LangChain makes it easy to build LLM powered applications"
```

```python
# SOLUTION

def stream_with_count(text: str):
    for i, word in enumerate(text.split(), start=1):
        print(f"[token {i:2}] {word}")
        yield word + " "

text = "LangChain makes it easy to build LLM powered applications"

result = ""
for token in stream_with_count(text):
    result += token

print(f"--- {len(text.split())} tokens streamed ---")
print(f"Full response: {result.strip()}")
```

## 8 — Everything Comes Together

>  *'Variables. Dicts. Functions. Classes. Async. Decorators. Generators. Watch them all show up in 15 lines that call a real AI.'*

```python
# pip install openai python-dotenv
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY", "sk-demo"))

# Function (Block 4) + Dict (Block 2) + Type hints (Block 4)
def ask(prompt: str, temperature: float = 0.7) -> str:
    response = client.chat.completions.create(
        model       = "gpt-4o",
        temperature = temperature,
        messages    = [
            {"role": "system", "content": "You are a concise AI tutor. Max 2 sentences."},
            {"role": "user",   "content": prompt},
        ]
    )
    return response.choices[0].message.content

print(ask("What is RAG?"))
```

```python
# Change temperature live — show the difference
print(" Factual (temp=0.1):")
print(ask("Explain LangChain", temperature=0.1))
print()
print(" Creative (temp=1.5):")
print(ask("Explain LangChain", temperature=1.5))
```

```python
# Generator + Streaming (Block 8) — the real thing
def ask_streaming(prompt: str):
    stream = client.chat.completions.create(
        model    = "gpt-4o",
        messages = [{"role": "user", "content": prompt}],
        stream   = True                    # ← enables generator mode
    )
    for chunk in stream:                   # ← for loop over generator
        token = chunk.choices[0].delta.content
        if token:
            yield token                    # ← yield, just like Block 8

print("Streaming: ", end="")
for token in ask_streaming("What is an AI agent in one sentence?"):
    print(token, end="", flush=True)
print()
```
